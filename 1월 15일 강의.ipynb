```python
import pandas as pd
mpg = pd.read_csv('C:/Users/82105/Downloads/mpg.csv')
mpg
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
## gridspec

import matplotlib.gridspec as gridspec
grid = gridspec.GridSpec(2,2) ## 그래프를 그릴 도화지 개수
plt.figure(figsize = (15,10))
plt.subplots_adjust(wspace = 0.4, hspace = 0.3) ## 그래프 간의 거리 조절

## 서버플롯 그리기
## sns countplot으로 그래프 그리기 (x에 대한 카운트를 그래프로 그리고 범례도 설정할 수 있다)
mpg_features = ['manufacturer', 'model', 'category', 'drv']
for idx, feature in enumerate(mpg_features):
    ax = plt.subplot(grid[idx])
    sns.countplot(x = feature, data = mpg, hue = 'drv', palette = 'pastel', ax = ax)
    ax.set_title(f'{feature} Distribution')
    
```
