```python
df = pd.DataFrame({'Math':[100,80,90], 'English':[60,50,70]}, index = ['A', 'B', 'C'])
```

```python
df
```

```python
sns.heatmap(df)
```

```python
sns.heatmap(df)
plt.xticks([0.5,1.5], [' ', '영어'])
```

```python
## 데이터 표시 annot True 박스와 대응되는 데이터 표시, fmt 데이터를 어떻게 표시할지 d 정수, .3f 소숫점 셋째

sns.heatmap(df, annot = True, fmt = 'd', linewidth = 1, linecolor = 'black')
```

```python
sns.heatmap(df, annot = True, fmt = 'd', linewidth = 1, linecolor = 'black', cmap = 'Blues')
```
