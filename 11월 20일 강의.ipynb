```python
# 람다식을 이용해서 데이터 전처리 해보기
```

```python
exam = pd.read_csv('C:/Users/82105/Downloads/exam.csv')
```

```python
exam.assign(tot = lambda x: x['math']+x['english']+x['science'])
```

```python
exam.assign(tot = lambda x: x['math']+x['english']+x['science'], mean = lambda x: x['tot']/3)
```

```python
# 집단별 요약(groupby)
- groupby('컬럼') # 해당 컬럼을 묶어서 보여준다.
- agg(변수 = ('컬럼', '원하는 통계치'))
```

```python
#반 별로 어떤 컬럼의 평균, 기타 통계치를 알고 싶다
exam.groupby('nclass').agg(mean_math = ('math', 'mean'))
```

```python
exam.groupby('nclass').agg(mean_math = ('math', 'sum'))
```

```python
# 여러 개의 변수의 요약 통계를 보고싶은 경우
# 반 별 영어, 수학, 과학 점수 평균, 합, 중위값, 등등

exam.groupby('nclass').mean() # 전체 요약 통계
```
